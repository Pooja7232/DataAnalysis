# SEABORN - Statistical Data Visualization

## What is Seaborn?

### -> Seaborn is a Python Library that builts on the  top of matplotlib that makes the statistical data more easy and beautiful.

## Why we need it?

### 1. For Creating plots like violin,Box and Pair.
### 2. Automatically Handles data from pandas dataframes.

## Where we use it?

### 1. Data Analysis and Reporting.
### 2. Exploring relationships between variables.
### 3. Machine Learning(For visualizing datasets and results)

```python
import seaborn as sns
```

```python
df = sns.load_dataset("titanic")
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
```

## Histograms & Countplots

### Histogram

```python
# Histogram
sns.histplot(df['age'].dropna(), bins=10, kde=True)
plt.title("Histogram of Age")
plt.show()
```

### Count Plot

```python
#Count Plot
sns.countplot(x='class', data=df)
plt.title("Count of Passengers by Class")
plt.show()
```

### Scatter Plot

```python
#Scatter Plot
sns.scatterplot(x='age', y='fare', data=df, hue='sex')
plt.title("Age vs Fare")
plt.show()
```

## Boxplot & Violin Plot

### Box Plot

```python
# Boxplot
sns.boxplot(x='class', y='age', data=df)
plt.title("Boxplot of Age by Class")
plt.show()
```

### Violin Plot

```python
# Violin plot
sns.violinplot(x='class', y='age', hue='sex', data=df, split=True)
plt.title("Violin Plot of Age by Class and Sex")
plt.show()
```

### Pairplot

```python
#Pairplot
iris = sns.load_dataset('iris')
sns.pairplot(iris, hue='species')
plt.show()
```

### Heat Map

```python
#Heatmap
# Correlation heatmap (numeric columns only)
corr = iris.select_dtypes(include=['float64', 'int64']).corr()
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Iris Dataset")
plt.show()
```

### Customized Histogram

```python
#Customization & Themes
sns.set_theme(style="whitegrid")  # options: darkgrid, white, ticks, etc.
sns.histplot(df['age'].dropna(), bins=10, color='orange')
plt.title("Customized Histogram with Seaborn")
plt.show()
```

### Bar Plot

```python
#Bar Plot
sns.barplot(x='class', y='fare', data=df, hue='sex', ci='sd')
plt.title("Average Fare by Class and Sex")
plt.show()
```

```python
#1.Compare Passenger Survival by Class and Gender
#Bar Plot
sns.barplot(x='class', y='survived', data=df, hue='sex')
plt.title("Comparing Passenger by Class and Gender")
plt.show()
```

```python
#2. Distribution of Passenger Ages
sns.histplot(df['age'].dropna(), bins=10,color="purple")
plt.title("Distribution of Passenger Ages")
plt.show()
```

```python
#3. Fare Distribution by Class
sns.boxplot(x='class', y='fare', data=df,color="yellow")
plt.title("Fare Distribution by Class")
plt.show()
```

```python
#4. Relationship Between Age and Fare
sns.scatterplot(x='age', y='fare', data=df,color="hotpink")
plt.title("Relationship between age and fare")
plt.show()
```

```python
#5. Correlation Heatmap
tit = sns.load_dataset('titanic')
corr = tit.select_dtypes(include=['float64', 'int64']).corr()
import seaborn as sns
import matplotlib.pyplot as plt
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap of Titanic Dataset")
plt.show()
```
